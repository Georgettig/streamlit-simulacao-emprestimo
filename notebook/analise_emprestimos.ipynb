```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../dados/emprestimos.csv")
print(df.head())
```

```python
print(df.describe())
```

```python
print(df["aprovado"].value_counts())
```

```python
sns.countplot(x="aprovado", data=df)
plt.title("Distribuição de Aprovação de Empréstimos")
plt.show()
```

```python
plt.hist(x = df['idade'])
plt.show()
```

```python
plt.hist(x = df['anos_trabalhados'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True)
plt.title("Correlação entre as variáveis")
plt.show()
```
